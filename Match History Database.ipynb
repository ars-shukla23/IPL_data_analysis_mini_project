Now we will be inputting the two teams.(Tkinter is not supported on jupyter hence I will input them manually for the jupyter file.)

```python
teams=np.array(['Chennai Super Kings',
'Deccan Chargers',
'Delhi Daredevils',
'Gujurat Lions',
'Kings XI Punjab',
'Kochi Tuskers Kerela',
'Kolkata Knight Riders',
'Mumbai Indians',
'Pune Warriors',
'Rajasthan Royals',
'Rising Pune Supergiants',
'Royal Challengers Bangalore',
'Sunrisers Hyderabad'])
m=1 #Array position of first team selection
n=7 #Array position of second team selection



```

Now we will be displaying the head-to-head fixtures and matches dataframes records for the selected teams.(in this case MI and DC)

```python
dfm.loc[((dfm['team1']==teams[m]) &(dfm['team2']==teams[n])) |((dfm['team1']==teams[n]) &(dfm['team2']==teams[m])),['id','date','venue','winner','result','result_margin','player_of_match']]
```

Now lets say if we had to display the records for a particular season. (We have created another dataframe called **dataframe_1** to prevent changes to original dataframe). But we notice that the 'date' column is not of 'datetime64' data type. hence we will change its data type to datetime64

```python
dataframe_1=dfm.loc[((dfm['team1']==teams[m]) &(dfm['team2']==teams[n])) |((dfm['team1']==teams[n]) &(dfm['team2']==teams[m])),['id','date','venue','winner','result','result_margin']]
dataframe_1['date']=dataframe_1['date'].apply(pd.to_datetime)
dataframe_1.dtypes
#We can clearly see that the datatype of date has been changed. 
```

Now coming back to displaying the data records for a particular season. Lets say 2011.

```python
season='2011'
if season=='2008':
    mask= (dataframe_1['date']>='2008-1-1') & (dataframe_1['date']<'2008-12-31')
elif season=='2009':
    mask= (dataframe_1['date']>='2009-1-1') & (dataframe_1['date']<'2009-12-31')
elif season=='2010':
    mask= (dataframe_1['date']>='2010-1-1') & (dataframe_1['date']<'2010-12-31')
elif season=='2011':
    mask= (dataframe_1['date']>='2011-1-1') & (dataframe_1['date']<'2011-12-31')
elif season=='2012':
    mask= (dataframe_1['date']>='2012-1-1') & (dataframe_1['date']<'2012-12-31')
elif season=='2013':
    mask= (dataframe_1['date']>='2013-1-1') & (dataframe_1['date']<'2013-12-31')
elif season=='2014':
    mask= (dataframe_1['date']>='2014-1-1') & (dataframe_1['date']<'2014-12-31')
elif season=='2015':
    mask= (dataframe_1['date']>='2015-1-1') & (dataframe_1['date']<'2015-12-31')
elif season=='2016':
    mask= (dataframe_1['date']>='2016-1-1') & (dataframe_1['date']<'2016-12-31')
elif season=='2017':
    mask= (dataframe_1['date']>='2017-1-1') & (dataframe_1['date']<'2017-12-31')
elif season=='2018':
    mask= (dataframe_1['date']>='2018-1-1') & (dataframe_1['date']<'2018-12-31')
elif season=='2019':
    mask= (dataframe_1['date']>='2019-1-1') & (dataframe_1['date']<'2019-12-31')
elif season=='2020':
    mask= (dataframe_1['date']>='2020-1-1') & (dataframe_1['date']<'2020-12-31')  
    
dataframe_1.loc[mask]
```

Here we will present the head-to-head performance of the selected teams graphically with the help of a pie chart.

```python
no_of_team1_wins=0
no_of_team2_wins=0
for i in dataframe_1['winner']:
    if i==teams[m]:
        no_of_team1_wins=no_of_team1_wins+1
    elif i==teams[n]:
        no_of_team2_wins=no_of_team2_wins+1
        
y=np.array([(no_of_team1_wins/(len(dataframe_1)))*100,(no_of_team2_wins/(len(dataframe_1)))*100])
labels1=["Matches won by "+teams[m],"Matches won by "+teams[n]]
plt.pie(y,labels=labels1)
plt.show()        
```

Here we will present toss win-loss record of the selected teams.

```python
no_of_team1_tosswins=0
no_of_team2_tosswins=0
for i in dfm['toss_winner']:
    if i==teams[m]:
        no_of_team1_tosswins=no_of_team1_tosswins+1
    elif i==teams[n]:
        no_of_team2_tosswins=no_of_team2_tosswins+1
        
y=np.array([(no_of_team1_tosswins/(len(dataframe_1)))*100,(no_of_team2_tosswins/(len(dataframe_1)))*100])
labels2=[" Tosses won by "+teams[m],"Tosses won by "+teams[n]]
plt.pie(y,labels=labels2)
plt.show()
```
