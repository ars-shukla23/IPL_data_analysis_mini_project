# Player Comparision

This section will be covering Player Comparision. Two players provided by the user will be compared on various aspects of their game such as runs scored, wickets taken etc.

We have imported our modules and have read the csv files using read_csv function.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dfm=pd.read_csv("ipl-matches.csv")
dfb=pd.read_csv("ipl-ball.csv")
```

List of all batsmen(unique)-

```python
batsman_list=np.array(dfb['batsman'])
unique_batsman_list=np.unique(batsman_list)
print(unique_batsman_list)
print("No of batsmen:",len(unique_batsman_list))
```

List of all bowlers(unique)-

```python
bowlers_list=np.array(dfb['bowler'])
unique_bowlers_list=np.unique(bowlers_list)
print(unique_bowlers_list)
print("No of bowlers:",len(unique_bowlers_list))
```

We start by declaring two vaiables for the players-player1 and player2.

```python
player1="DJ Bravo"
player2="HH Pandya"
batsman_cond1=dfb['batsman']==player1
bowler_cond1=dfb['bowler']==player1
batsman_cond2=dfb['batsman']==player2
bowler_cond2=dfb['bowler']==player2
player1_df=dfb.where(batsman_cond1 | bowler_cond1)
player2_df=dfb.where(batsman_cond2 | bowler_cond2)
```

We change the datatypes of batsman_runs, is_wicket, extra_runs to int from object using astype function. Also from the dataframes player1_df and player2_df that we have created we create new dataframes without NaN values(i.e: newplayer1_df and newplayer2_df)

```python
newplayer1_df=player1_df.dropna(axis=0,how="all")
newplayer2_df=player2_df.dropna(axis=0,how="all")

newplayer1_df=newplayer1_df.astype({"batsman_runs":'int',"is_wicket":'int',"extra_runs":'int'})
newplayer2_df=newplayer2_df.astype({"batsman_runs":'int',"is_wicket":'int',"extra_runs":'int'})
```

**Runs Comparision-** We compare the runs of the two players.

```python
player1_runs=newplayer1_df['batsman_runs'].sum()
player2_runs=newplayer2_df['batsman_runs'].sum()
print("The total runs scored by",player1,"is",player1_runs)
print("The total runs scored by",player2,"is",player2_runs)
```

**Runs Comparision Graph-** We draw a bar plot for graphical representation

```python
players = [player1,player2]
runs= [player1_runs,player2_runs]
fig = plt.figure(figsize = (8, 4))
plt.bar(players,runs,color ='orange',width = 0.2)
plt.grid(color ='black',linestyle ='-.', linewidth = 0.8,alpha = 0.4) 
plt.xlabel("The name of selected players")
plt.ylabel("Runs scored")
plt.title("Runs comparision of the two Players")
plt.show() 
```

**Wickets Comparision-** We compare the wickets of the two players

```python
player1_wickets=newplayer1_df['is_wicket'].sum()
player2_wickets=newplayer2_df['is_wicket'].sum()
print("The total wickets taken by",player1,"is",player1_wickets)
print("The total wickets taken by",player2,"is",player2_wickets)
```

**Wickets Comparison Graph-** We draw a bar plot for graphical representation.

```python
players = [player1,player2]
wickets=[player1_wickets,player2_wickets]
fig = plt.figure(figsize = (8, 4))
plt.bar(players,wickets,color ='purple',width = 0.2)
plt.grid( color ='black',linestyle ='-.', linewidth = 0.8,alpha = 0.4) 
plt.xlabel("The name of selected players")
plt.ylabel("Wickets taken")
plt.title("Wickets comparision of the two Players")
plt.show() 

```

**Fielding(Catches) Comparision**- We compare the number of catches taken by the players. 

```python
catches_player1=0
catches_player2=0
for i in dfb['fielder']:
    if i==player1:
        catches_player1+=1
    elif i==player2:
        catches_player2+=1
    else:
        pass
print("The number of catches taken by",player1,"are",catches_player1)
print("The number of catches taken by",player2,"are",catches_player2) 
```

**Fielding Comparision Graph-**

```python
players = [player1,player2]
catches=[catches_player1,catches_player2]
fig = plt.figure(figsize = (8, 4))
plt.bar(players,catches,color ='blue',width = 0.2)
plt.grid( color ='black',linestyle ='-.', linewidth = 0.8,alpha = 0.4) 
plt.xlabel("The name of selected players")
plt.ylabel("Catches taken")
plt.title("Fielding comparision of the two Players")
plt.show() 
```
