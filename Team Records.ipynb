**Total matches won-** Displays total matches won by each team.

```python

print(dfm['winner'].value_counts(dropna=True))
winner_array=np.array(dfm['winner'].value_counts(dropna=False))
print(winner_array)
```

Clearly some data cleaning must be performed here(both Rising Pune Supergiants and Rising Pune Supergiant exist)

```python
for x in dfm.index:
    if dfm.loc[x, 'winner']=="Rising Pune Supergiant":
        dfm.loc[x,'winner'] = "Rising Pune Supergiants"       
print(dfm['winner'].value_counts(dropna=True))        
 #Now that the data has been cleaned we can proceed       
```

**Pie-chart for total matches won-**

```python
label1=["Mumbai Indians","Chennai Super Kings","Kolkata Knight Riders","Royal Challengers Bangalore","Kings XI Punjab","Rajasthan Royals","Delhi Daredevils","Sunrisers Hyderabad","Deccan Chargers","Delhi Capitals","Rising Pune Supergiants","Gujurat Lions","Pune Warriors","Kochi Tuskers Kerala","Tie/NR"]
plt.pie(winner_array,labels=label1,shadow=True)

plt.show()
```

**Bar-Chart for total matches won-**

```python
x=label1
y=winner_array
plt.barh(x , y ,color="teal")
plt.show()
```
